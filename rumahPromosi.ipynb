```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv("rumahPromosi.csv")
x = df['x'].values
y = df['y'].values
```

```python
slope, intercept, r, p, std_err = stats.linregress(x,y)

print('slope = ', slope)
print('intercept = ', intercept)
print('r = ', r)
```

```python
def myfunc(x):
    return slope * x + intercept
```

```python
mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
print('Penjualan rumah jika biaya promosi = 160 yaitu sebesar', myfunc(160))
```
