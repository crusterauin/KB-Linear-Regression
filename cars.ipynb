```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import matplotlib as mpl
```

```python
df = pd.read_csv("cars.csv")
reg = LinearRegression()
reg.fit(df[['volume','weight']].values,df.co2)
```

```python
mpl.rcParams['legend.fontsize'] = 12
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(df.volume, df.weight, df.co2)
ax.legend()
ax.view_init(45, 0)
plt.show()
```

```python
a = reg.coef_
b = reg.intercept_

X_baru = np.array([1000, 790]).reshape(-1, 2)
prediksi_Y = reg.predict(X_baru)

print('Koefisien regresi = ', a[0], ' dan ', a[1])
print('Intercept = ', b)
print('Hasil prediksi untuk x1=1000 dan x2=790 yaitu ', prediksi_Y)
```
